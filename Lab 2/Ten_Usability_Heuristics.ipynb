##1. Visibility of system status

The system should always keep users informed about what is going on, through appropriate feedback within reasonable time.

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image1.png?raw=true)

It is easy to tell if the buttons on adobe.com are working as the visual feedback is easy to follow. When pushing buttons visual feedback is given to the user in different colors. For example, the buttons underneath the menu, the usually grey words will become light blue, and a light grey square will appear around the selected button. 

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image5.png?raw=true)

When there is more time required to finish the task when pushing a button, a blue/grey circle will appear indicating to the user that there is waiting time involved.

##2. Match between system and the real world

The system should speak the users' language, with words, phrases and concepts familiar to the user, rather than system-oriented terms. Follow real-world conventions, making information appear in a natural and logical order.

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image7.png?raw=true)

There is a match between the system and the real world as Adobe offers various methods of payment for different kinds of software. They reflect and closely replicate the monthly subscription for cell phone subscription fees with varying kinds of access depending on how much you are willing to pay. As it is monthly, you are also free to cancel your subscription after a month of use much like your cell phone bills. This makes payment easier and more understandable for users.

##3. User control and freedom

Users often choose system functions by mistake and will need a clearly marked "emergency exit" to leave the unwanted state without having to go through an extended dialogue. Support undo and redo.

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image3.png?raw=true)

The Adobe homepage has a cancel button for some purposes but not everything. For example, when you are signing in and you have made a mistake, there is a cancel button. However, if you were to search a term up on the webpage and wanted to go back after clicking through links, there is not a button for that action. The user would have to rely on their internet browser to provide them with an undo button, rather than use one that was provided by Adobe.

##4. Consistency and standards
Users should not have to wonder whether different words, situations, or actions mean the same thing. Follow platform conventions.

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image2.png?raw=true)

Adobe has strong consistency on its primary webpage. As shown, the site follows the convention for keeping dropdown menus along the top of the site, and has a search feature, as well as a link to a sign-in page, in the upper lefthand corner. Throughout the home page, it emphasizes “action items” by using blue buttons instead of regular hypertext links. However, once you navigate to the page to, for example, download Adobe Reader, a free application, you end up on a completely different site that does not have the same toolbars or style conventions.

##5. Error prevention

Even better than good error messages is a careful design which prevents a problem from occurring in the first place. Either eliminate error-prone conditions or check for them and present users with a confirmation option before they commit to the action.

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image12.png?raw=true)

As seen above in the Adobe Account password management page, there is a field that requires you to confirm your current account password before changing it. This prevents you from accidentally changing your password or another user who may have malicious intent. In addition, if you happen to leave your account logged in on a public computer it prevents account information being changed. Having this check point is a prime example of error prevention for the user.

##6. Recognition rather than recall

Error messages should be expressed in plain language (no codes), precisely indicate the problem, and constructively suggest a solution.

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image6.png?raw=true)

Revisiting the Adobe password management page, we see a few errors. These errors are highlighted in red as well as where the error is located. This is a great example of recognition as the user only has to look and see where things are red to find the errors. In addition, the reason for the error is in plain text the user can read rather than an error code. This is considered recognition because rather than just having the reasons for the error in red it also has the text field where the error exists. If the text field was instead not colors this would require the user to recall where these errors are relevant and then fix their password.

##7. Flexibility and efficiency of use

Accelerators -- unseen by the novice user -- may often speed up the interaction for the expert user such that the system can cater to both inexperienced and experienced users. Allow users to tailor frequent actions.

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image9.png?raw=true)

Adobe has been known for their use of shortcuts to ease the process in their apps. As for their online website apps, while there are less of them and not all of them are advertised they are there. As seen in the right hand side there is a (A) which means you can use A key to either lock/unlock the aspect ratio. As for the non-advertised as shown the crop function is enabled, users can use the error keys to crop this picture in a much finer way than with the mouse.

##8. Aesthetic and minimalist design

Dialogues should not contain information which is irrelevant or rarely needed. Every extra unit of information in a dialogue competes with the relevant units of information and diminishes their relative visibility.

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image10.png?raw=true)

As seen here on the home page once you login there is very minimal information. It is clean and simple leaving you with all the information you need. You can see your apps and services, your subscription, your profile, and well as access to documentation/get help. It has the key items without going to in depth or containing more information than needed.

##9. Help users recognize, diagnose, and recover from errors

Error messages should be expressed in plain language (no codes), precisely indicate the problem, and constructively suggest a solution.

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image4.png?raw=true)

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image8.png?raw=true)

Adherence to the heuristic of recognizing, diagnosing, and recovering from errors is also good. The user receives visual feedback and clear directions if they input an invalid email address on the login page (other forms also adhere to this). If the user tries to navigate to a page that doesn’t exist, the 404 page explains the problem (page not found) and offers solutions (return to homepage or search the website) in a simple, straightforward manner. While the main heading of the 404 page is a bit ambiguous, there is plain language immediately following to provide clear messaging.

##10. Help and documentation

Even though it is better if the system can be used without documentation, it may be necessary to provide help and documentation. Any such information should be easy to search, focused on the user's task, list concrete steps to be carried out, and not be too large.

![link text](https://github.com/andrewixl/IST-250/blob/master/Lab%202/images/image11.png?raw=true)

Adobe.com has an excellent collection of high-quality documentation and help articles. In addition to a strong search feature with helpful suggestions as you type, the main help page provides links to specific applications you may be looking for answers to, as well as a list of popular help articles broken up into categories. Using the search takes the user to a search engine-like page listing articles by relevance. The articles themselves rely heavily on screenshots for step-by-step instructions, and some include video tutorials embedded in the article. If the user cannot find a relevant answer, the site also provides a community support forum broken out by application, so users can search there, or post their own question.
